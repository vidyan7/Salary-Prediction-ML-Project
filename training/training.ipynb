```python
df.dropna(inplace=True)
```

```python
df['Gender'].unique()
df['Gender']=df['Gender'].map({'Male':1,'Female':0,'Other':2})
df['Gender']
```

```python
df['Education Level'] = df['Education Level'].map({
    "High School": 0,
    "Bachelor's": 1,
    "Bachelor's Degree": 1,
    "Master's": 2,
    "Master's Degree": 2,
    "PhD": 3,
    "phD":3
})
df['Education Level']
```

```python
df['Job Title'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
# creating object
lb_Job_Title=LabelEncoder()
print(df['Job Title'].unique())
# transform data using object method
df['Job Title']=lb_Job_Title.fit_transform(df['Job Title'])
df['Job Title'].unique()
```

```python
df
```

```python
y=df['Salary']
y.unique() 

x=df.drop('Salary', axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
# calling the function for split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(x_train,y_train)
```

```python
result=LR.predict(x_test)
result
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error
mae=mean_absolute_error(y_test,result)
print("mean_absolute_error:",mae)
mse=mean_absolute_error(y_test,result)  
print("MSE",mse)
rmse=root_mean_squared_error(y_test,result)
print("RMSE",rmse)
```

```python
train_score=LR.score(x_train,y_train)
test_score=LR.score(x_test,y_test)
print("LR model train score",train_score)
print("LR model test score",test_score)
```

# Decision Tree

```python
train_score=DT.score(x_train,y_train)
test_score=DT.score(x_test,y_test)
print("LR model train score",train_score)
print("LR model test score",test_score)
```

# RandomForest Classifier

```python
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor()
RF.fit(x_train, y_train)
```

```python
rf_result = RF.predict(x_test)
rf_result
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error
mae=mean_absolute_error(y_test,rf_result)
print("mean_absolute_error:",mae)
mse=mean_absolute_error(y_test,rf_result)  
print("MSE",mse)
rmse=root_mean_squared_error(y_test,rf_result)
print("RMSE",rmse)
```

```python
train_score=RF.score(x_train,y_train)
test_score=RF.score(x_test,y_test)
print("LR model train score",train_score)
print("LR model test score",test_score)
```

# Deployment

```python
import pickle
with open("salary_rf_model.pkl","rb")as f:
    rf_model=pickle.load(f)
with open("lb_Job_Title.pkl","rb")as f:
    lb_Job_Title=pickle.load(f)
```

```python
def predict_salary(age=34.0,Gender='Male',Education_Level='PhD',Job_Title='Senior Data Engineer',Years_of_Experience=5.0):
    lst=[]
    lst=lst+[age]
    if Gender=="Female":
        lst=lst+[0]
    elif Gender=="Male":
        lst=lst+[1]
    elif Gender=="Other":
        lst=lst+[2]

    if Education_Level=="High School":
        lst=lst+[0]
    elif Education_Level=="Bachelor's Degree":
        lst=lst+[1]
    elif Education_Level==" Master's Degree":
        lst=lst+[2]
    elif Education_Level=="PhD":
        lst=lst+[3]

    Job_Title=lb_Job_Title.transform([Job_Title])
    lst=lst+list(Job_Title)
    lst=lst+[Years_of_Experience]
    print(lst)
    result=rf_model.predict([lst])
    return result[0]
    
```

```python
age = float(input("Enter age: "))
Gender = input("Enter gender (Male/Female/Other): ")
Education_Level = input("Enter education level (High School/Bachelor's Degree/ Master's Degree/PhD): ")
Job_Title = input("Enter job title (must match trained model job titles): ")
Years_of_Experience = float(input("Enter years of experience: "))

prediction=predict_salary(age,Gender,Education_Level,Job_Title,Years_of_Experience)
print("Predict Salary is:",prediction)
```
